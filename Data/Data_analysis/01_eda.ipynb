```python
import pandas as pd

df_fake = pd.read_csv("/content/drive/MyDrive/AI_projects/Fake_News_Detector/dataset/Fake.csv")
df_real = pd.read_csv("/content/drive/MyDrive/AI_projects/Fake_News_Detector/dataset/True.csv")

df_fake["label"] = 0
df_real["label"] = 1

df = pd.concat([df_fake, df_real])
df.head()
```

```python
df["label"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["text_length"], bins=50)
plt.title("Distribution of Text Length")
plt.xlabel("Text Length")
plt.ylabel("Count")
```
